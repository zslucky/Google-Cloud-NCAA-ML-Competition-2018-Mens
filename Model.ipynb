```python
df_teams = pd.read_csv('./dataset/DataFiles/Teams.csv')
df_teams_rename_colums = {'TeamID': 'team_id'}
df_teams_drop_columns = ['TeamName', 'FirstD1Season', 'LastD1Season']
df_teams = df_teams.rename(columns=df_teams_rename_colums)
df_teams = df_teams.drop(df_teams_drop_columns, axis=1)
df_teams = df_teams.astype(dtype='int')
df_teams = df_teams.set_index('team_id')

df_nacc_tourney_compact_results = pd.read_csv('./dataset/DataFiles/NCAATourneyCompactResults.csv')
df_nacc_tourney_compact_results.head()
```

```python
df_ncaa_win_teams = df_nacc_tourney_compact_results.groupby(['Season', 'WTeamID'])['WScore'].mean().reset_index().rename(columns={'Season': 'season', 'WTeamID': 'team_id', 'WScore': 'team_score'})
df_ncaa_lose_teams = df_nacc_tourney_compact_results.groupby(['Season', 'LTeamID'])['LScore'].mean().reset_index().rename(columns={'Season': 'season', 'LTeamID': 'team_id', 'LScore': 'team_score'})
```

```python
df_ncaa_win_teams.head()
```

```python
df_ncaa_lose_teams.head()
```

```python
df_ncaa_teams = pd.concat([df_ncaa_win_teams, df_ncaa_lose_teams])
df_ncaa_teams = df_ncaa_teams.groupby(['season', 'team_id'])['team_score'].mean().reset_index()
df_ncaa_teams.head()
```
